```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import os
```

Define Dataset Path and Commands


```python
# Path to the dataset
DATA_PATH = 'speech_commands'

# Define target commands (you can modify this list)
COMMANDS = ['yes', 'no', 'up', 'down', 'left', 'right', 'on', 'off', 'stop', 'go']

print("Target commands:", COMMANDS)
```

Load and Visualize an Audio File



```python
# Load a sample file
sample_file = os.path.join(DATA_PATH, 'yes', '0a7c2a8d_nohash_0.wav')
audio, sr = librosa.load(sample_file, sr=16000)  # Resample to 16 kHz

# Display the waveform
plt.figure(figsize=(10, 4))
librosa.display.waveshow(audio, sr=sr)
plt.title("Waveform of the audio signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

print(f"Sample rate: {sr}")
print(f"Audio duration: {len(audio) / sr} seconds")
```

Normalize Audio Length

```python
def pad_or_trim(audio, sr=16000, duration=1.0):
    max_length = int(sr * duration)
    if len(audio) > max_length:
        return audio[:max_length]
    else:
        return np.pad(audio, (0, max_length - len(audio)), mode='constant')

# Normalize the audio length
normalized_audio = pad_or_trim(audio)
print(f"Normalized audio length: {len(normalized_audio)} samples")
```

 Listen to the Audio

```python
from IPython.display import Audio

# Play the audio file
Audio(normalized_audio, rate=sr)
```

**Extracting MFCC Features**

```python
def extract_mfcc(audio, sr=16000, n_mfcc=40):
    """
    Extract MFCC features from an audio signal.
    Parameters:
        audio: Normalized audio signal
        sr: Sampling rate
        n_mfcc: Number of MFCC features to extract
    Returns:
        MFCC features as a 2D NumPy array
    """
    mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)
    return mfcc

# Example: Extract MFCCs from the normalized audio
mfcc_features = extract_mfcc(normalized_audio)
print(f"MFCC shape: {mfcc_features.shape}")
```

Visualizing MFCCs

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc_features, x_axis='time', sr=sr, cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title("MFCCs")
plt.xlabel("Time (s)")
plt.ylabel("MFCC Coefficients")
plt.show()
```

- Extracting Mel-Spectrograms

```python
def extract_mel_spectrogram(audio, sr=16000, n_mels=128):
    """
    Extract Mel-Spectrogram from an audio signal.
    Parameters:
        audio: Normalized audio signal
        sr: Sampling rate
        n_mels: Number of Mel bands
    Returns:
        Mel-Spectrogram as a 2D NumPy array
    """
    mel_spec = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=n_mels)
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)  # Convert to dB scale
    return mel_spec_db

# Example: Extract Mel-Spectrogram
mel_spectrogram = extract_mel_spectrogram(normalized_audio)
print(f"Mel-Spectrogram shape: {mel_spectrogram.shape}")

# Visualize the Mel-Spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(mel_spectrogram, x_axis='time', y_axis='mel', sr=sr, cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title("Mel-Spectrogram")
plt.xlabel("Time (s)")
plt.ylabel("Mel Frequency Bands")
plt.show()
```

- Batch Feature Extraction

```python
def process_dataset(data_path, commands, feature_type='mfcc', n_features=40):
    """
    Extract features for all audio files in the dataset.
    Parameters:
        data_path: Root path of the dataset
        commands: List of target commands
        feature_type: 'mfcc' or 'mel_spectrogram'
        n_features: Number of features to extract (MFCC or Mel bands)
    Returns:
        A list of features and their corresponding labels
    """
    features = []
    labels = []
    for label, command in enumerate(commands):
        command_path = os.path.join(data_path, command)
        for file in os.listdir(command_path):
            file_path = os.path.join(command_path, file)
            audio, _ = librosa.load(file_path, sr=16000)
            audio = pad_or_trim(audio)

            if feature_type == 'mfcc':
                feature = extract_mfcc(audio, n_mfcc=n_features)
            elif feature_type == 'mel_spectrogram':
                feature = extract_mel_spectrogram(audio, n_mels=n_features)

            features.append(feature)
            labels.append(label)

    return np.array(features), np.array(labels)

# Example: Process dataset for MFCC features
X, y = process_dataset(DATA_PATH, COMMANDS, feature_type='mfcc', n_features=40)
print(f"Extracted features shape: {X.shape}")
print(f"Labels shape: {y.shape}")
```

**Preprocessing Features and Splitting the Dataset**

- Reshaping Features for Consistency

```python
# Define a fixed size for feature frames (e.g., 40 MFCCs x 100 time frames)
FIXED_SHAPE = (40, 100)

def reshape_features(features, fixed_shape=FIXED_SHAPE):
    """
    Reshape features to have a consistent size.
    Parameters:
        features: List of feature arrays
        fixed_shape: Desired shape (e.g., 40x100 for MFCCs)
    Returns:
        Reshaped features as a NumPy array
    """
    reshaped_features = []
    for feature in features:
        if feature.shape[1] < fixed_shape[1]:
            # Pad time frames
            pad_width = fixed_shape[1] - feature.shape[1]
            feature = np.pad(feature, ((0, 0), (0, pad_width)), mode='constant')
        else:
            # Truncate time frames
            feature = feature[:, :fixed_shape[1]]
        reshaped_features.append(feature)
    return np.array(reshaped_features)

# Apply reshaping
X = reshape_features(X)
print(f"Reshaped features shape: {X.shape}")
```

- Normalize Features

```python
from sklearn.preprocessing import StandardScaler

# Flatten features for normalization
scaler = StandardScaler()
X_flattened = X.reshape(X.shape[0], -1)
X_normalized = scaler.fit_transform(X_flattened)

# Reshape back to original shape
X_normalized = X_normalized.reshape(X.shape)
print(f"Normalized features shape: {X_normalized.shape}")
```

- Split Dataset into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42, stratify=y)

print(f"Training features shape: {X_train.shape}, Labels shape: {y_train.shape}")
print(f"Testing features shape: {X_test.shape}, Labels shape: {y_test.shape}")
```

- Convert to Tensor Format

```python
import tensorflow as tf

# Convert to TensorFlow datasets
X_train_tf = tf.convert_to_tensor(X_train, dtype=tf.float32)
y_train_tf = tf.convert_to_tensor(y_train, dtype=tf.int32)

X_test_tf = tf.convert_to_tensor(X_test, dtype=tf.float32)
y_test_tf = tf.convert_to_tensor(y_test, dtype=tf.int32)

print("Training tensors created!")
```

**Building the CNN-RNN Model Architecture**

```python
from tensorflow.keras import layers, models

def build_cnn_rnn_model(input_shape, num_classes):
    """
    Builds a CNN-RNN model for speech command recognition.
    Parameters:
        input_shape: Shape of the input (e.g., 40x100x1 for MFCCs)
        num_classes: Number of target commands
    Returns:
        Compiled Keras model
    """
    model = models.Sequential()

    # Input layer
    model.add(layers.Input(shape=input_shape))

    # CNN layers
    model.add(layers.Conv2D(32, (3, 3), activation='relu', padding='same'))
    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D((2, 2)))

    model.add(layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D((2, 2)))

    # Reshape for RNN layers
    model.add(layers.Reshape((-1, 64)))  # Flatten time-frequency map to sequence of 64-dimensional vectors

    # RNN layers
    model.add(layers.GRU(128, return_sequences=True))
    model.add(layers.GRU(128))

    # Attention mechanism (optional but recommended)
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dropout(0.3))

    # Output layer
    model.add(layers.Dense(num_classes, activation='softmax'))

    return model

# Specify input shape and number of classes
input_shape = (40, 100, 1)  # 40 MFCCs x 100 time frames, single channel
num_classes = len(COMMANDS)

# Build and compile the model
model = build_cnn_rnn_model(input_shape, num_classes)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

-  Reshape Input Data

```python
# Add channel dimension for Conv2D input
X_train_reshaped = X_train[..., np.newaxis]  # Shape: (samples, 40, 100, 1)
X_test_reshaped = X_test[..., np.newaxis]    # Shape: (samples, 40, 100, 1)

print(f"Reshaped training data: {X_train_reshaped.shape}")
```

- Compile the Model

```python
model.compile(
    optimizer='adam',  # Or use AdamW for improved performance
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

- Train Model

```python
history = model.fit(
    X_train_reshaped, y_train,
    validation_data=(X_test_reshaped, y_test),
    epochs=10,
    batch_size=32,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True),
        tf.keras.callbacks.ReduceLROnPlateau(patience=2)
    ]
)
```

- Evaluate the model

```python
test_loss, test_accuracy = model.evaluate(X_test_reshaped, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
def predict_command(file_path):
    """
    Predicts the command from an audio file.

    Parameters:
        file_path (str): Path to the audio file.

    Returns:
        str: Predicted command.
    """
    if not os.path.exists(file_path):
        print(f"Error: File not found at {file_path}")
        return None

    try:
        audio, sr = librosa.load(file_path, sr=16000)
    except (soundfile.LibsndfileError, FileNotFoundError, RuntimeError) as e:
        print(f"Error loading audio file: {e}")
        return None

    # Extract features and reshape for model input
    mfcc = extract_mfcc(audio)
    mfcc = reshape_features([mfcc])[0]  # Reshape using your existing function
    mfcc = mfcc[np.newaxis, ..., np.newaxis]  # Add batch and channel dimensions

    # Make prediction
    prediction = model.predict(mfcc)
    predicted_label = COMMANDS[np.argmax(prediction)]

    return predicted_label

# Example usage: Predict a new command
file_path = 'speech_commands/yes/0a7c2a8d_nohash_0.wav'
predicted_command = predict_command(file_path)

if predicted_command:
    print(f"Predicted Command: {predicted_command}")
else:
    print("Prediction failed due to file error.")
```

```python
import threading  # For background processing

# Load your trained model
model = load_model('speech_command_model.keras')

# Parameters for audio recording
fs = 16000  # Sample rate
duration = 1  # Duration of recording in seconds

# Global variable to store prediction results
prediction_result = None

def predict_from_microphone():
    """Records audio from the microphone, extracts features, and predicts the command."""
    global prediction_result  # Access the global variable

    print("Say a command...")

    # Get a list of available audio devices
    devices = sd.query_devices()
    print("Available audio devices:")
    for i, device in enumerate(devices):
        print(f"{i}: {device['name']}")

    # Select the input device (you might need to change the index)
    selected_device_index = int(input("Enter the index of the input device to use: "))

    # Record audio using the selected device
    audio_data = sd.rec(int(duration * fs), samplerate=fs, channels=1, device=selected_device_index)
    sd.wait()  # Wait until recording is finished

    # Extract MFCC features
    mfcc = extract_mfcc(audio_data.flatten(), sr=fs)  # Assuming you have your extract_mfcc function
    mfcc = reshape_features([mfcc])[0]  # Reshape using your existing function
    mfcc = mfcc[np.newaxis, ..., np.newaxis]  # Add batch and channel dimensions

    # Predict the command
    prediction = model.predict(mfcc)
    predicted_label = COMMANDS[np.argmax(prediction)]

    # Store the prediction result in the global variable
    prediction_result = predicted_label

    print(f"Predicted command: {predicted_label}")


def run_prediction_loop():
    """Runs the prediction loop in a separate thread."""
    while True:
        predict_from_microphone()
        time.sleep(1)  # Add a small delay before the next prediction
```

```python
if __name__ == "__main__":
    # Start the prediction loop in a background thread
    prediction_thread = threading.Thread(target=run_prediction_loop)
    prediction_thread.daemon = True  # Allow the main thread to exit
    prediction_thread.start()

    # You can add other logic here for your main application,
    # e.g., display prediction results in a GUI.
    while True:
        if prediction_result is not None:
            print("Main application: Predicted command:", prediction_result)
            # Process the prediction result as needed in your main application
            prediction_result = None
        time.sleep(0.1)
```
